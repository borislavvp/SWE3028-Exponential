```python
import tensorflow as ft
import os
import pandas as pd
import seaborn
import plotly


primeFrame = pd.read_csv('./SPY.csv')

print(primeFrame)
```

```python
primeFrame = primeFrame.dropna() 
primeFrame = primeFrame.rename(columns= {'Datetime': 'date', 'Close': 'close', 'Open': 'open', 'High': 'high', 'Low': 'low', 'Volume': 'volume'}) 
primeFrame['date'] = pd.to_datetime(primeFrame['date']) 
primeFrame = primeFrame.sort_values(by=['date'], ascending=True) 
print(primeFrame.iloc[100:200])
primeFrame = primeFrame.iloc[100:200]
```

```python
import matplotlib.pyplot as plt
# 필요한 모듈 import 하기 
import plotly
import plotly.graph_objects as go
import plotly.express as px

# %matplotlib inline 은 jupyter notebook 사용자용 - jupyter notebook 내에 그래프가 그려지게 한다.
%matplotlib inline 
```

```python
import plotly.graph_objects as go

import pandas as pd
from datetime import datetime
```

```python
fig = go.Figure(data=[go.Candlestick(x=primeFrame['date'],
                open=primeFrame['open'],
                high=primeFrame['high'],
                low=primeFrame['low'],
                close=primeFrame['close'])])

fig.show()
```

```python
import plotly.graph_objects as go

trace = go.Heatmap(z=primeFrame['volume'],
                   x=primeFrame['date'],
                   y=primeFrame['close'])
data=[trace]
layout = go.Layout(title='Stock Price with Activity Heatmap of Volume')

figure = go.Figure(data=data, layout=layout)

f2 = go.FigureWidget(figure)
f2.write_image ('./graph.jpg') #특정 파일 형식을 지정할 수 있다.
#plotly.offline.plot(f2, filename='./graph.html') ##웹 형태로 저장된다.
f2
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.express as px

fig2 = make_subplots(specs=[[{"secondary_y": True}]])

fig2.add_trace(go.Candlestick(x=primeFrame['date'],
                open=primeFrame['open'], high=primeFrame['high'],
                low=primeFrame['low'], close=primeFrame['close'],
                name='CandleStick'),
                secondary_y= True
                )

# include a go.Bar trace for volumes
fig2.add_trace(go.Bar(x=primeFrame['date'], y=primeFrame['volume'],
                name='Volume',marker_color = '#00008B',marker_opacity = 0.2),
               secondary_y=False)

fig2.layout.yaxis2.showgrid=False

fig2.update_layout(
    showlegend=False,
    xaxis_rangeslider_visible=False
)

fig2.show()
```
